```python
# Randomly generates a binary search tree with values from 0 up to 9, and displays it growing up.
#
# Written by *** and Eric Martin for COMP9021


import sys
from random import seed, choice
from binary_tree_adt import *

from collections import defaultdict
def print_growing_up(tree):
    height = tree.height()
    length = 2**(height + 1) - 1
    d = defaultdict(list)
    for i in range(height+1):
        for j in range(length):
            d[i].append(' ')
    root = 2**height - 1
    d[0][root] = tree.value
    if tree.left_node:
        add_num(tree.left_node, 0, 0, root, height, length, d)
    if tree.right_node:
        add_num(tree.right_node, 0, 1, root, height, length, d)
#     print(d)
    list0 = []
    for i in range(height + 1):
        j = len(d[i])
        for k in range(j-1,-1,-1):
            if d[i][k]==' ' and d[i][k-1]!=' ':
                list0.append(k)
                break
    # print(list0)
    for i in range(height+1):
        if list0: 
            j = list0[i]
            k = len(d[i]) - 1
            if d[i][k] == ' ':
                while k >= j:
                    del d[i][k]
                    k = k-1
    for i in range(height,-1,-1):
        for j in d[i]:
            print(j, end='')
        print()
    print(d)
def add_num(tree, level, child, place, height, length, d):
    nlevel = level + 1
    width = 2**(height - nlevel)
    if child == 0:
        nplace = place - width
    else:
        nplace = place + width
    if tree.value is not None:
        d[nlevel][nplace] = tree.value
    if tree.left_node:
        add_num(tree.left_node, nlevel, 0, nplace, height, length, d)
    if tree.right_node:
        add_num(tree.right_node, nlevel, 1, nplace, height, length, d)

# Possibly write additional function(s)
        

try:
    seed_arg, nb_of_nodes = (int(x) for x in
                              input('Enter two integers, with the second one between 0 and 10: '
                                   ).split()
                            )
    if nb_of_nodes < 0 or nb_of_nodes > 10:
        raise ValueError
except ValueError:
    print('Incorrect input, giving up.')
    sys.exit()

seed(seed_arg)
data_pool = list(range(nb_of_nodes))
tree = BinaryTree()
for _ in range(nb_of_nodes):
    datum = choice(data_pool)
    tree.insert_in_bst(datum)
    data_pool.remove(datum)
print_growing_up(tree)
           
```
